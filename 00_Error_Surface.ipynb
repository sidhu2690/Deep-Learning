```python
import numpy as np
import matplotlib.pyplot as plt

x1, y1 = 0.5, 0.2
x2, y2 = 2.5, 0.9

def loss_fn(w, b):
  return 0.5 * ((y1 - (w*x1 + b))**2 + (y2 - (w*x2 + b))**2)

# ranges for w and b (Guess)
w_values = np.linspace(-3, 3, 50)
b_values = np.linspace(-3, 3, 50)

# Compute the loss values for the guessed  and 
W, B, L = [], [], []
for w in w_values:
    for b in b_values:
        loss = loss_fn(w, b)
        W.append(w)
        B.append(b)
        L.append(loss)

W, B, L = np.array(W), np.array(B), np.array(L)

fig = plt.figure(figsize=(6, 10))
ax = plt.axes(projection="3d")

ax.scatter(0.5, 0, loss_fn(0.5, 0), color="red", s=50)
ax.scatter(-0.1, 0, loss_fn(-0.1, 0), color="red", s=50)

ax.plot_trisurf(W, B, L, cmap="viridis")
ax.set_xlabel("w")
ax.set_ylabel("b")
ax.set_zlabel("L", labelpad = -3)
plt.show()
```

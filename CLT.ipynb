```python
import matplotlib.pyplot as plt
import random
import math

mu = 5.0

# Generate exponential random variable
def expo():
    return -mu * math.log(1 - random.random())
```

```python
n_values = []
means = []
samples = []

for i in range(1, 1001):
    x = expo()
    samples.append(x)
    if i % 10 == 0:
        n_values.append(i)
        means.append(sum(samples) / len(samples))


plt.plot(n_values, means, 'b-', linewidth=2)
plt.axhline(mu, color='r', linestyle='--', label=f'μ={mu}')
plt.xlabel('Sample Size (n)')
plt.ylabel('Sample Mean')
plt.title('Law of Large Numbers')
plt.legend()
plt.grid(True, alpha=0.3)
```

```python
sample_means = []
for i in range(1000):
    sample = [expo() for j in range(30)]
    sample_means.append(sum(sample) / 30)

plt.hist(sample_means, bins=30, density=True, alpha=0.7, edgecolor='black')
plt.axvline(mu, color='r', linestyle='--', label=f'μ={mu}')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.title('Central Limit Theorem (n=30)')
plt.grid(True, alpha=0.3)
```
